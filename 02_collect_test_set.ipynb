#### 셀 1: 라이브러리 import

```python
# 경로 처리 라이브러리
from pathlib import Path

# 파일 복사 라이브러리
import shutil

# 랜덤 샘플링 라이브러리
import random

# JSON 파일 처리 (사용 기록 저장용)
import json

# 날짜/시간 라이브러리
from datetime import datetime

print("✅ 라이브러리 import 완료")
```

#### 셀 2: 경로 설정 및 폴더 생성

```python
# ============================================
# Test set 894장 수집
# (화재 447장 + 정상 447장)
# ============================================

# 프로젝트 루트 경로
PROJECT_ROOT = Path(r'N:\개인\이수빈\3.13_Mini_Project')

# 소스 경로 (끝난 프로젝트)
SOURCE_FIRE = Path(r'N:\개인\이수빈\끝난프로젝트\끝_project_fire\Data\fire')
SOURCE_NORMAL = Path(r'N:\개인\이수빈\끝난프로젝트\끝_project_fire\Data\normal')

# Test set 저장 경로
TEST_DIR = PROJECT_ROOT / 'DATASET' / 'test_set_894'
TEST_FIRE = TEST_DIR / 'fire'
TEST_NORMAL = TEST_DIR / 'normal'

# 폴더 생성
TEST_FIRE.mkdir(parents=True, exist_ok=True)  # 화재 폴더 생성
TEST_NORMAL.mkdir(parents=True, exist_ok=True)  # 정상 폴더 생성

print("=" * 60)
print("📦 Test Set 수집 (894장)")
print("=" * 60)
print(f"\n저장 경로: {TEST_DIR}")
```

#### 셀 3: 소스 데이터 확인

```python
# ============================================
# 소스 데이터 확인 (.webp 포함)
# ============================================

print("\n[소스 데이터 확인]")

# 화재 폴더의 모든 파일 가져오기
all_fire_files = list(SOURCE_FIRE.iterdir())
# 이미지 파일만 필터링 (.webp 포함!)
fire_images = [f for f in all_fire_files 
               if f.suffix.lower() in ['.jpg', '.jpeg', '.png', '.webp']]

# 정상 폴더의 모든 파일 가져오기
all_normal_files = list(SOURCE_NORMAL.iterdir())
# 이미지 파일만 필터링 (.webp 포함!)
normal_images = [f for f in all_normal_files 
                 if f.suffix.lower() in ['.jpg', '.jpeg', '.png', '.webp']]

print(f"  화재:  {len(fire_images)}개")
print(f"  정상:  {len(normal_images)}개")

# 확장자별 개수 확인
print(f"\n[확장자별 분포]")
fire_exts = {}
for img in fire_images:
    ext = img.suffix.lower()
    fire_exts[ext] = fire_exts.get(ext, 0) + 1

print(f"  화재:")
for ext, count in sorted(fire_exts.items()):
    print(f"    {ext}: {count}개")

normal_exts = {}
for img in normal_images:
    ext = img.suffix.lower()
    normal_exts[ext] = normal_exts.get(ext, 0) + 1

print(f"  정상:")
for ext, count in sorted(normal_exts.items()):
    print(f"    {ext}: {count}개")

# 충분한지 확인
if len(fire_images) < 447:
    print(f"\n⚠️ 화재 이미지 부족: {len(fire_images)}개 (필요: 447개)")
else:
    print(f"\n✅ 화재 이미지 충분: {len(fire_images)}개 (필요: 447개)")

if len(normal_images) < 447:
    print(f"⚠️ 정상 이미지 부족: {len(normal_images)}개 (필요: 447개)")
else:
    print(f"✅ 정상 이미지 충분: {len(normal_images)}개 (필요: 447개)")
```

#### 셀 4: 랜덤 샘플링

```python
# ============================================
# 랜덤 샘플링 (447장씩)
# ============================================

print("\n[랜덤 샘플링]")

# 랜덤 시드 고정 (재현 가능)
random.seed(42)

# 화재 447장 선택
if len(fire_images) >= 447:
    selected_fire = random.sample(fire_images, 447)  # 447장 랜덤 선택
    print(f"  화재:  447장 선택 ✅")
else:
    selected_fire = fire_images  # 전부 선택
    print(f"  ⚠️ 화재 부족: {len(fire_images)}장만 사용")

# 정상 447장 선택
if len(normal_images) >= 447:
    selected_normal = random.sample(normal_images, 447)  # 447장 랜덤 선택
    print(f"  정상:  447장 선택 ✅")
else:
    selected_normal = normal_images  # 전부 선택
    print(f"  ⚠️ 정상 부족: {len(normal_images)}장만 사용")
```

#### 셀 5: 화재 이미지 복사

```python
# ============================================
# 화재 이미지 복사
# ============================================

print("\n[화재 이미지 복사 중...]")

# 복사 기록 리스트 (사용 기록용)
copied_fire = []

# 각 이미지 복사
for idx, img_path in enumerate(selected_fire, 1):
    # 목표 파일명 생성 (중복 방지)
    dst_path = TEST_FIRE / f"fire_test_{idx:04d}{img_path.suffix}"
    
    # 파일 복사
    shutil.copy2(img_path, dst_path)
    
    # 사용 기록에 추가
    copied_fire.append({
        'original': str(img_path),  # 원본 경로
        'copied_to': str(dst_path),  # 복사된 경로
        'filename': img_path.name  # 파일명
    })
    
    # 진행 상황 출력 (50장마다)
    if idx % 50 == 0:
        print(f"  {idx}/{len(selected_fire)}")

print(f"  ✅ 화재 {len(copied_fire)}장 완료")
```

#### 셀 6: 정상 이미지 복사

```python
# ============================================
# 정상 이미지 복사
# ============================================

print("\n[정상 이미지 복사 중...]")

# 복사 기록 리스트
copied_normal = []

# 각 이미지 복사
for idx, img_path in enumerate(selected_normal, 1):
    # 목표 파일명 생성
    dst_path = TEST_NORMAL / f"normal_test_{idx:04d}{img_path.suffix}"
    
    # 파일 복사
    shutil.copy2(img_path, dst_path)
    
    # 사용 기록에 추가
    copied_normal.append({
        'original': str(img_path),  # 원본 경로
        'copied_to': str(dst_path),  # 복사된 경로
        'filename': img_path.name  # 파일명
    })
    
    # 진행 상황 출력 (50장마다)
    if idx % 50 == 0:
        print(f"  {idx}/{len(selected_normal)}")

print(f"  ✅ 정상 {len(copied_normal)}장 완료")
```

#### 셀 7: 사용 기록 저장

```python
# ============================================
# 사용 기록 JSON 저장
# ============================================

print("\n[사용 기록 저장]")

# 사용 기록 데이터 구성
usage_record = {
    'created_at': datetime.now().strftime('%Y-%m-%d %H:%M:%S'),  # 생성 시각
    'purpose': 'Test set for final evaluation (70:15:15)',  # 목적
    'total_count': len(copied_fire) + len(copied_normal),  # 총 개수
    'fire_count': len(copied_fire),  # 화재 개수
    'normal_count': len(copied_normal),  # 정상 개수
    'source': {  # 소스 정보
        'fire': str(SOURCE_FIRE),
        'normal': str(SOURCE_NORMAL)
    },
    'destination': str(TEST_DIR),  # 목적지
    'fire_files': copied_fire,  # 화재 파일 상세 목록
    'normal_files': copied_normal  # 정상 파일 상세 목록
}

# JSON 파일로 저장
record_path = TEST_DIR / 'test_set_usage_record.json'
with open(record_path, 'w', encoding='utf-8') as f:
    json.dump(usage_record, f, indent=2, ensure_ascii=False)  # 한글 깨짐 방지

print(f"  ✅ 사용 기록: {record_path}")
```

#### 셀 8: 최종 확인 및 비율 계산

```python
# ============================================
# 최종 확인 (.webp, .jpeg 포함!)
# ============================================

print(f"\n{'='*60}")
print("📊 Test Set 수집 완료")
print(f"{'='*60}")

# 실제 복사된 파일 개수 확인 (모든 확장자!)
final_fire = [f for f in TEST_FIRE.iterdir() 
              if f.suffix.lower() in ['.jpg', '.jpeg', '.png', '.webp']]
final_normal = [f for f in TEST_NORMAL.iterdir() 
                if f.suffix.lower() in ['.jpg', '.jpeg', '.png', '.webp']]

# 저장 위치 출력
print(f"\n저장 위치: {TEST_DIR}")

# 파일 개수 출력
print(f"\n파일 수:")
print(f"  화재:  {len(final_fire)}장")
print(f"  정상:  {len(final_normal)}장")
print(f"  총:    {len(final_fire) + len(final_normal)}장")

# ============================================
# 전체 데이터 비율 계산
# ============================================

print(f"\n{'='*60}")
print("📊 전체 데이터 비율 (Train+Val+Test)")
print(f"{'='*60}")

# 학습 데이터 개수
train_count = 4172
val_count = 894
test_count = len(final_fire) + len(final_normal)
total_all = train_count + val_count + test_count

# 비율 계산
train_ratio = train_count / total_all * 100
val_ratio = val_count / total_all * 100
test_ratio = test_count / total_all * 100

# 출력
print(f"\n전체: {total_all:,}장")
print(f"  Train: {train_count:,}장 ({train_ratio:.1f}%)")
print(f"  Val:   {val_count:,}장 ({val_ratio:.1f}%)")
print(f"  Test:  {test_count:,}장 ({test_ratio:.1f}%)")

print(f"\n비율: {train_ratio:.0f}:{val_ratio:.0f}:{test_ratio:.0f}")
print(f"목표: 70:15:15")

# 목표 달성 확인
if abs(train_ratio - 70) < 1 and abs(val_ratio - 15) < 1 and abs(test_ratio - 15) < 1:
    print("\n✅ 목표 비율 달성!")
else:
    print(f"\n⚠️ 비율 차이:")
    print(f"  Train: {train_ratio - 70:+.1f}%p")
    print(f"  Val:   {val_ratio - 15:+.1f}%p")
    print(f"  Test:  {test_ratio - 15:+.1f}%p")

print(f"\n{'='*60}")
print("✅ 완료")
print(f"{'='*60}")
```
