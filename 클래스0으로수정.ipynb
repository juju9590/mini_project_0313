```python
import os
import tkinter as tk
from tkinter import filedialog

def unify_yolo_classes():
    # 1. 라벨 폴더 선택
    root = tk.Tk()
    root.withdraw()
    root.attributes('-topmost', True)
    
    print("📂 클래스를 수정할 라벨(.txt) 폴더를 선택하세요...")
    label_dir = filedialog.askdirectory(title="라벨 폴더 선택")
    
    if not label_dir:
        print("❌ 폴더가 선택되지 않았습니다.")
        return

    # 2. .txt 파일 목록 가져오기
    label_files = [f for f in os.listdir(label_dir) if f.lower().endswith('.txt')]
    
    if not label_files:
        print("⚠️ 해당 폴더에 .txt 파일이 없습니다.")
        return

    print(f"🚀 총 {len(label_files)}개의 파일을 수정합니다...")

    changed_count = 0
    for filename in label_files:
        file_path = os.path.join(label_dir, filename)
        
        # 파일 읽기
        with open(file_path, 'r') as f:
            lines = f.readlines()
        
        new_lines = []
        for line in lines:
            parts = line.split()
            if len(parts) > 0:
                # 맨 앞의 클래스 번호만 '0'으로 교체하고 나머지 좌표는 그대로 유지
                parts[0] = '0'
                new_lines.append(" ".join(parts))
        
        # 수정된 내용 다시 저장
        with open(file_path, 'w') as f:
            f.write("\n".join(new_lines))
        
        changed_count += 1
        if changed_count % 100 == 0:
            print(f"⏳ 진행 중... ({changed_count}/{len(label_files)})")

    print("-" * 30)
    print(f"✨ 작업 완료! 모든 클래스가 '0'으로 변경되었습니다.")

if __name__ == "__main__":
    unify_yolo_classes()
```
