```python
import os
import glob
import tkinter as tk
from tkinter import filedialog
from ultralytics import YOLO

# 1. 폴더 선택 창 띄우기 (이미지가 들어있는 'images' 폴더 선택)
def select_folder():
    root = tk.Tk()
    root.withdraw()
    root.attributes('-topmost', True)
    
    print("📂 이미지가 들어있는 폴더(예: images)를 선택하세요...")
    folder_selected = filedialog.askdirectory(title="이미지 폴더를 선택해주세요")
    
    root.destroy()
    return folder_selected

# 폴더 선택 실행
image_folder = select_folder()

if not image_folder:
    print("❌ 폴더가 선택되지 않았습니다. 프로그램을 종료합니다.")
    exit()

# 2. 경로 및 모델 설정
# 선택한 image_folder의 부모 디렉토리에 'labels' 폴더 생성
parent_dir = os.path.dirname(image_folder)
label_folder = os.path.join(parent_dir, 'labels')

# 라벨 저장 폴더 생성 (이미지 폴더와 같은 레벨)
if not os.path.exists(label_folder):
    os.makedirs(label_folder)
    print(f"📁 새 폴더 생성됨: {label_folder}")

# 모델 로드
model_path = '../전처리후비교/best_rhyzen.pt.pt'  # 모델 경로
print("🤖 AI 모델을 불러오는 중...")
model = YOLO(model_path)

# 3. 이미지 파일 목록 가져오기
image_extensions = ['*.jpg', '*.jpeg', '*.png', '*.bmp', '*.webp']
image_files = []
for ext in image_extensions:
    image_files.extend(glob.glob(os.path.join(image_folder, ext)))

if not image_files:
    print("⚠️ 폴더 내에 이미지 파일이 없습니다.")
    exit()

print(f"🚀 총 {len(image_files)}개의 파일을 분석합니다.")
print(f"📍 저장 위치: {label_folder}")

# 4. 루프를 돌며 분석 및 라벨 생성
for img_path in image_files:
    file_name = os.path.splitext(os.path.basename(img_path))[0]
    
    # AI 분석
    results = model.predict(source=img_path, conf=0.3, verbose=False)
    
    yolo_data = []
    for r in results:
        for box in r.boxes:
            class_id = 0 # 클래스 번호 0번 고정
            
            # YOLO 포맷 좌표 추출
            xywhn = box.xywhn[0].tolist()
            line = f"{class_id} {xywhn[0]:.6f} {xywhn[1]:.6f} {xywhn[2]:.6f} {xywhn[3]:.6f}"
            yolo_data.append(line)
    
    # -------------------------------------------------------
    # TXT 파일 저장 (image_folder와 나란한 labels 폴더에 저장)
    # -------------------------------------------------------
    label_path = os.path.join(label_folder, f"{file_name}.txt")
    with open(label_path, 'w', encoding='utf-8') as f:
        f.write("\n".join(yolo_data))

print(f"\n✨ 작업 완료!")
print(f"✅ 최종 이미지 위치: {image_folder}")
print(f"✅ 최종 라벨 위치: {label_folder}")
```
